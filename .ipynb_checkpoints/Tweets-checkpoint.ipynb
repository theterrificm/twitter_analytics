```python
import json
import pandas as pd

f = open('./data/tweets.json', encoding="utf8")
data = json.load(f)
this_dict = {
    "fav_count" : [],
    "rt_count" : [],
    "date" : [],
    "id": []
}
for x in range(len(data)):
    #     this_dict["url"].append( data[x]['tweet']['entities']['urls'][0]['url']) 
    this_dict["fav_count"].append( data[x]['tweet']['favorite_count'])
    this_dict["rt_count"].append(data[x]['tweet']['retweet_count'])
    this_dict["date"].append( data[x]['tweet']['created_at'])
    this_dict["id"].append( data[x]['tweet']['id'])



f.close()
    
```

```python
dataframe = {'fav': this_dict['fav_count'],
        'rt_count': this_dict['rt_count'],
        'date': this_dict['date'],
        'id' : this_dict['id']
       }
df= pd.DataFrame(dataframe)
```

```python
df.head()
```

# Adding Year, Month, Day and Hour Columns

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_week'] = df['date'].dt.dayofweek
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df.head(30)
```
